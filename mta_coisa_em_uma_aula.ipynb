```python
# Entrada dos números
num1 = int(input("Diga um número: "))
num2 = int(input("Diga outro número: "))

# Operações
soma = num1 + num2
subtracao = num1 - num2
multiplicacao = num1 * num2
divisao = num1 / num2
divisao_inteira = num1 // num2
modulo = num1 % num2
potencia = num1 ** num2

# Exibindo os resultados
print(f"{num1} mais {num2} é igual a {soma}, "
      f"{num1} menos {num2} é igual a {subtracao}, "
      f"{num1} vezes {num2} é igual a {multiplicacao}, "
      f"{num1} dividido por {num2} é igual a {divisao}, "
      f"{num1} dividido por {num2} (divisão inteira) é igual a {divisao_inteira}, "
      f"O resto da divisão de {num1} por {num2} é igual a {modulo}, "
      f"{num1} elevado a {num2} é igual a {potencia}")
```
